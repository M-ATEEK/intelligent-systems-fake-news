```python
chunks = []
chunk_size = 10000
rows_collected = 0
target_rows = 300000

for chunk in pd.read_csv(
    "../data/raw/news_cleaned_2018_02_13.csv",
    chunksize=chunk_size,
    engine="python",
    on_bad_lines="skip",
    encoding="utf-8",
    encoding_errors="replace",
):
    chunks.append(chunk)
    rows_collected += len(chunk)
    if rows_collected >= target_rows:
        break

df_small = pd.concat(chunks, ignore_index=True).head(target_rows)
print(f"Loaded {len(df_small):,} rows")

df_small.to_csv("../data/processed/news_small.csv", index=False)
print("Saved to ../data/processed/news_small.csv")
```
